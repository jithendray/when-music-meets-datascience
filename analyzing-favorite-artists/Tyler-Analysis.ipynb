```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')

import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
tyler = pd.read_csv('tyler.csv')
tyler.head()
```

```python
df = tyler.drop(['id','uri', 'artist'], axis=1)
```

```python
df['album'] = pd.Categorical(df['album'], ['Goblin', 'Wolf', 'Cherry Bomb', 'Flower Boy', 'IGOR'])
```

```python
plt.figure(figsize=(8,5))
df.groupby(by='album')['album'].count().plot(kind='bar')
plt.title('Number of songs in each album!')
plt.show()
```

### acousticness

```python
plt.figure(figsize=(8,7))
plt.subplot(2, 1, 1)
df.groupby(by='album')['acousticness'].mean().plot()
plt.grid(b=None)
plt.subplot(2, 1, 2)
df.groupby(by='album')['acousticness'].mean().plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10,4)
plt.subplot(1, 3, 1)
plt.xlabel('Aucosticness')
plt.title('Goblin, Wolf')
df[df.album.isin(['Goblin', 'Wolf'])].acousticness.plot.hist(bins=10, ylim=[0,20], range=(0,1));
plt.subplot(1, 3, 2)
plt.xlabel('Aucosticness')
plt.title('Cherry Bomb')
df[df.album.isin(['Cherry Bomb'])].acousticness.plot.hist(bins=10, range=(0,1), ylim=[0,20])
plt.subplot(1, 3, 3)
plt.xlabel('Aucosticness')
plt.title('Flower Boy, IGOR')
df[df.album.isin(['Flower Boy', 'IGOR'])].acousticness.plot.hist(bins=10, range=(0,1), ylim=[0,20])
```

### Distribution of energy in each album for every song

```python
plt.figure(figsize=(10,7))
plt.plot(df[df['album'] == 'Goblin']['energy'].values, label='Goblin')
plt.plot(df[df['album'] == 'Wolf']['energy'].values, label='Wolf')
plt.plot(df[df['album'] == 'Cherry Bomb']['energy'].values, label='Cherry Bomb')
plt.plot(df[df['album'] == 'Flower Boy']['energy'].values, label='Flower Boy')
plt.plot(df[df['album'] == 'IGOR']['energy'].values, label='IGOR', color='#ff0392')
plt.grid(b=None)
plt.title('Distribution of energy in for every song each album')
plt.legend(loc='best')
plt.show()
```

### Distribution of acousticness in for every song each album
