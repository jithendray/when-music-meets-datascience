# Analysing Black Sabbath Discography

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
bs = pd.read_csv('black-sabbath.csv')
bs.head()
```

```python
bs.columns
```

```python
df = bs.drop(['track_number','id','uri', 'artist'], axis=1)
df.columns
```

```python
df.groupby(by='album')['album'].count()
```

```python
df.album.unique()
```

```python
df['album'] = pd.Categorical(df['album'], list(df.album.unique()))
```

```python
df.groupby(by='album')['album'].count()
```

__Features extracted:__

- acousticness
- danceability
- energy
- instrumentalness
- liveness
- loudness
- speechiness
- tempo
- valence
- popularity

```python
df.groupby(by='album')['album'].count().plot(kind='bar')
plt.grid(b=None)
plt.show()
```

```python
f, axes = plt.subplots(2, 1)
ax = sns.lineplot(x="album", y="acousticness", data=df, ax=axes[0], sort=False)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax = sns.barplot(x="album", y="acousticness", data=df, ax=axes[1])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```
